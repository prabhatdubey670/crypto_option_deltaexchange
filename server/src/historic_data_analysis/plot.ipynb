```python
import pandas as pd 
from matplotlib.pyplot import plot, show

path = 'D:/codeputs/ongoing/crypto_option_deltaexchange/server/data/EVENTS_ANALYSIS.xlsx'

df = pd.read_excel(path, sheet_name='USA_BTC', index_col='Date', parse_dates=True)
df2 = pd.read_excel(path, sheet_name='BTC_5Y', index_col='Date', parse_dates=True)

# Assuming you want to join based on the index ('Date')
df = df.join(df2)

# Filter the DataFrame for a specific date range


# Select specific columns
df_filtered = df[['Event', 'BTC% Change ']]
df_filtered.loc[df_filtered['BTC% Change '] == 100, 'BTC% Change '] = 0


df_filtered.plot()
show()
```
